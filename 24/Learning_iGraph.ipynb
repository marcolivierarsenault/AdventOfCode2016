```r
createsquare4NeigboursList <- function(n, m){
    return <- numeric()
    for(i in 1:(n)){
        for(j in 1:(m)){
            current <- ((i-1)*n)+j
            if(j<(m))
                return <- c(return,current,(current+1))
            if(i<(n))
                return <- c(return,current,(current+n))
        } 
    }
    return
}
q <- createsquare4NeigboursList(3,3)
q
```

```r
edges <- q
g <- graph(edges, n=max(edges), directed=FALSE)
plot(g)
```
